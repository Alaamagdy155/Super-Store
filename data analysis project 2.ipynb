```python
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')

plt.figure(figsize=(12, 6))
plt.scatter(df['Discount'], df['Profit'], alpha=0.6, color='green', label='Profit vs Discount')
plt.title('Relationship Between Profit and Discount', fontsize=16)
plt.xlabel('Discount', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.axhline(0, color='red', linestyle='--', linewidth=1)  
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.regplot(data=df, x='Discount', y='Sales', scatter=True, color='purple', line_kws={"color": "red"})
plt.title('Sales vs Discount with Trendline', fontsize=16)
plt.xlabel('Discount', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.regplot(data=df, x='Sales', y='Profit', scatter=True, color='orange', line_kws={"color": "red"})
plt.title('Sales vs Profit with Trendline', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
